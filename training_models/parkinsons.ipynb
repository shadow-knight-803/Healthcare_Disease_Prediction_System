```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE  # Import SMOTE for balancing dataset
import pickle
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['name_encoded'] = le.fit_transform(df['name'])  # Convert names to numeric values
df = df.drop(columns=['name'])  # Drop original name column if not needed

df.head()
```

```python
X = df.drop(columns=['status'], axis=1)
y = df['status']
```

```python
X.head()
```

```python
X.columns = X.columns.astype(str)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split dataset into training and testing sets (30% test size for better evaluation)
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Apply SMOTE with controlled sampling strategy (not full balance)
smote = SMOTE(sampling_strategy=0.8, random_state=42)
X_train_sm, Y_train_sm = smote.fit_resample(X_train, Y_train)
```

```python
from sklearn.linear_model import LogisticRegression # Import LogisticRegression
# Baseline Model - Logistic Regression for comparison
baseline_clf = LogisticRegression()
baseline_clf.fit(X_train_sm, Y_train_sm)
y_pred_baseline = baseline_clf.predict(X_test)

print("\nBaseline Model (Logistic Regression) Results:")
print("Accuracy:", accuracy_score(Y_test, y_pred_baseline))
print("Precision:", precision_score(Y_test, y_pred_baseline))
print("Recall:", recall_score(Y_test, y_pred_baseline))
print("F1-score:", f1_score(Y_test, y_pred_baseline))
```

```python
# Hyperparameter Tuning with GridSearchCV for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 20],
    'class_weight': ['balanced']
}

grid = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='recall')
grid.fit(X_train_sm, Y_train_sm)
```

```python
# Best Model Selection
best_model = grid.best_estimator_
print("\nBest Model Parameters:", grid.best_params_)
```

```python
# Predict probabilities for threshold tuning
y_pred_proba = best_model.predict_proba(X_test)[:, 1]
```

```python
# Adjust decision threshold (default is 0.5, we lower it to increase recall)
threshold = 0.45  # Lowering it increases recall
y_pred = (y_pred_proba >= threshold).astype(int)
```

```python
# Model Evaluation
accuracy = accuracy_score(Y_test, y_pred)
precision = precision_score(Y_test, y_pred)
recall = recall_score(Y_test, y_pred)
f1 = f1_score(Y_test, y_pred)

# Display results
print("\nRandom Forest Model Results:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

# Classification Report and Confusion Matrix
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))

# Perform Cross-validation for better model evaluation
cv_scores = cross_val_score(best_model, X_scaled, y, cv=5, scoring='accuracy')
print("\nCross-validated Accuracy:", np.mean(cv_scores))
```

```python
# Save Model & Scaler
with open("parkinsons_model.sav", "wb") as model_file:
    pickle.dump(best_model, model_file)

with open("parkinsons_scaler.sav", "wb") as scaler_file:
    pickle.dump(scaler, scaler_file)
```
